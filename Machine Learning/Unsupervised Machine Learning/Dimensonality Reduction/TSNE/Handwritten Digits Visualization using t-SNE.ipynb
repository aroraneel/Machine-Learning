```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
X = pd.DataFrame(digits.data)
y = digits.target
```

```python
np.random.seed(7)
idx = np.random.choice(len(X), 1200, replace=False)
X_sample = X.iloc[idx]
y_sample = y[idx]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_sample)
```

```python
from sklearn.manifold import TSNE
tsne_20 = TSNE(
    n_components=2,
    perplexity=20,
    learning_rate=200,
    max_iter=1000,
    random_state=42
)
X_tsne_20 = tsne_20.fit_transform(X_scaled)
```

```python
from sklearn.manifold import TSNE
tsne_40 = TSNE(
    n_components=2,
    perplexity=40,
    learning_rate=200,
    max_iter=1000,
    random_state=42
)
X_tsne_40 = tsne_40.fit_transform(X_scaled)
```

```python
print(X_tsne_20.shape, X_tsne_40.shape)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(
    X_tsne_20[:, 0],
    X_tsne_20[:, 1],
    c=y_sample,
    cmap='tab10',
    s=25,
    alpha=0.8
)
plt.title('t-SNE (Perplexity = 20)')
plt.xlabel('Dim 1')
plt.ylabel('Dim 2')

plt.subplot(1, 2, 2)
plt.scatter(
    X_tsne_40[:, 0],
    X_tsne_40[:, 1],
    c=y_sample,
    cmap='tab10',
    s=25,
    alpha=0.8
)
plt.title('t-SNE (Perplexity = 40)')
plt.xlabel('Dim 1')
plt.ylabel('Dim 2')

plt.tight_layout()
plt.show()
```
