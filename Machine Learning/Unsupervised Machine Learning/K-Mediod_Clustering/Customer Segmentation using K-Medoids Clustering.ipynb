```
from sklearn_extra.cluster import KMedoids
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```
df = pd.read_csv('/content/CC GENERAL.csv')
df.head()
```

```
df.info()
```

```
df.tail()
```

```
df.shape
```

```
df = df.drop(columns=['CUST_ID'])
```

```
df.isnull().sum()
```

```
print(df.var().sort_values())
```

```
x = df[['CASH_ADVANCE_TRX','PURCHASES_TRX','INSTALLMENTS_PURCHASES',
        'ONEOFF_PURCHASES','BALANCE','CASH_ADVANCE',
        'PURCHASES','MINIMUM_PAYMENTS','PAYMENTS','CREDIT_LIMIT']]
```

```
x
```

```
x[['MINIMUM_PAYMENTS','CREDIT_LIMIT']] = x[['MINIMUM_PAYMENTS','CREDIT_LIMIT']].fillna(
    x[['MINIMUM_PAYMENTS','CREDIT_LIMIT']].mean()
)
```

```
x.isnull().sum()
```

```
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
```

```
x_scaled
```

```
kmedoids = KMedoids(n_clusters=5, random_state=42)
```

```
kmedoids.fit(x_scaled)
```

```
labels = kmedoids.labels_
```

```
labels
```

```
df['Cluster'] = labels
```

```
df.head(20)
```

```
df['Cluster'].value_counts()
```

```
import matplotlib.pyplot as plt

df['Cluster'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.title('Cluster Distribution (K-Medoids)')
plt.show()
```
