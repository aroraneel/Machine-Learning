```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"A:\my_codes\Machine Learning\Supervised Machine Learning\Classification\Logistic Algorithm\Personality Prediction\personality_dataset.csv")
df
```

```python
le = LabelEncoder()
df['Personality'] = le.fit_transform(df['Personality'])
df.head()
```

```python
plt.scatter(df.Time_spent_Alone, df.Social_event_attendance,df.Personality, color = 'red', marker = 'o')
```

```python
df.shape
```

```python
X = df[['Time_spent_Alone', 'Social_event_attendance']]
y = df.Personality
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=35)
```

```python
X_test
```

```python
X_train
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score (X_test, y_test)
```

```python
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
```

```python
plt.plot(fpr, tpr, color='blue', label='ROC Curve')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
auc = roc_auc_score(y_test, y_pred_proba)
print("AUC Score:", auc)
```
