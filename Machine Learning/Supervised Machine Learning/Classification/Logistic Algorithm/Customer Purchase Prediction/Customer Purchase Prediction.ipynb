```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
df = pd.read_csv(r"A:\my_codes\Machine Learning\Supervised Machine Learning\Classification\Logistic Algorithm\Customer Purchase Prediction\customer_purchase_data.csv")
df
```

```python
# Create scatter plot of Age vs Time Spent on Website, colored by Purchase Status
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df.Age, 
                     df.TimeSpentOnWebsite,
                     cmap='coolwarm',
                     alpha=0.6)
plt.colorbar(scatter, label='Purchase Status')
plt.xlabel('Age')
plt.ylabel('Time Spent on Website')
plt.title('Customer Age vs Time Spent by Purchase Status')
plt.show()
```

```python
df.shape
```

```python
X = df[['Age', 'TimeSpentOnWebsite']]
y = df.PurchaseStatus
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=35)
```

```python
X_test
```

```python
y_test
```

```python
X_train
```

```python
y_test
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
```

```python
plt.plot(fpr, tpr, color='blue', label='ROC Curve')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
auc = roc_auc_score(y_test, y_pred_proba)
print("AUC Score:", auc)
```
