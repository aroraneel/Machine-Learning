```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"A:\my_codes\my_project\Machine Learning\Supervised Machine Learning\Regression\Linear Regression\BMW Engine Size Prediction\bmw.csv",encoding='unicode_escape') 
df
```

```python
x = df[['year','price','mileage','tax','mpg']]
y = df[['engineSize']]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# Check for NaN values in the original data
print("NaN values in df:")
print(df.isnull().sum())
print("\nNaN values in x:")
print(x.isnull().sum())
print("\nNaN values in y:")
print(y.isnull().sum())
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
print('accuracy (R2):',accuracy)
```
