```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"A:\my_codes\my_project\Machine Learning\Supervised Machine Learning\Regression\Linear Regression\Medical Insurance Cost Prediction\insurance.csv",encoding='unicode_escape')
df
```

```python
df['sex'] = df['sex'].map({'male': 1, 'female': 0})
df['smoker'] = df['smoker'].map({'yes': 1, 'no': 0})
df['region'] = df['region'].map({'southeast': 0, 'southwest': 1, 'northeast': 2, 'northwest': 3})
df
```

```python
x = df[['age','sex','bmi','children','smoker','region']]
y = df['charges']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
print('accuracy (R2):',accuracy)
```
