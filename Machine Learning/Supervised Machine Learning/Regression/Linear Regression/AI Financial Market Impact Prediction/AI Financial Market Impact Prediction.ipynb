```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"A:\my_codes\Machine Learning\Supervised Machine Learning\Regression\Linear Regression\AI Financial Market Impact Prediction\ai_financial_market_daily_realistic_synthetic.csv",encoding='unicode_escape')
df
```

```python
df.drop (['Date', 'Event'], axis=1, inplace=True)
```

```python
df['Company'].unique()
```

```python
df['Company'] = df['Company'].map({'OpenAI': 0, 'Meta ': 1, 'Google': 2, 'Microsoft': 3})
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
x = df[[ 'R&D_Spending_USD_Mn', 'AI_Revenue_USD_Mn',
       'AI_Revenue_Growth_%','Stock_Impact_%']]
y  = df[['Company']]
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score

accuracy = r2_score(y_test, y_pred)
print("R2:", accuracy)
```
