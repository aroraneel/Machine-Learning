```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
d1 = pd.read_csv(r'c:\Users\Neel Arora\Downloads\Student_Performance.csv',encoding='unicode_escape')
d1
```

```python
d1.drop(columns='Extracurricular Activities', inplace=True)
d1

```

```python
x = d1[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
y= d1[['Performance Index']]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
print('accuracy (R2):',accuracy)
```
