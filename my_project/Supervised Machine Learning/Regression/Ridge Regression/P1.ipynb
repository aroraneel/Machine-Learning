```python
from sklearn.datasets import load_diabetes
import pandas as pd

data = load_diabetes()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

df

```

```python
df.columns
```

```python
X = df[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']]

Y= df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import Ridge

Ridge_CV = Ridge(alpha=1.0)
Ridge_CV.fit(X_train, Y_train)

Ridge_CV
```

```python
X_train
```

```python
Y_train
```

```python
y_pred = Ridge_CV.predict(X_test)
y_pred
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.metrics import r2_score

accuracy = r2_score(Y_test, y_pred)
print("Accuracy (R²):", accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

plt.scatter(Y_test, y_pred, label="Predictions")

max_val = max(max(Y_test), max(y_pred))
min_val = min(min(Y_test), min(y_pred))

plt.plot([min_val, max_val], [min_val, max_val],
         color="red", linewidth=2, label="Perfect Fit")

plt.xlabel("Actual Bike Rentals")
plt.ylabel("Predicted Bike Rentals")
plt.title("Actual vs Predicted Bike Usage (Ridge Regression)")
plt.legend()
plt.show()
```
