```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
d3 = pd.read_csv(r"A:\my_codes\my_project\Supervised Machine Learning\Regression\Linear Regression\AI Financial Market Impact Prediction\ai_financial_market_daily_realistic_synthetic.csv",encoding='unicode_escape')
d3
```

```python

d3.drop(('Date'), axis=1 ,inplace=True)
d3
```

```python
x = d3[['R&D_Spending_USD_Mn','AI_Revenue_USD_Mn','AI_Revenue_Growth_%',]]
y = d3[['Stock_Impact_%']]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
print('accuracy (R2):',accuracy)
```
