```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\Neel Arora\Downloads\personality_datasert.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
data['Personality'] = le.fit_transform(data['Personality'])

data.head()
```

```python

plt.scatter(data.Time_spent_Alone, data.Social_event_attendance,data.Personality, color = 'red', marker = 'o')
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X = data[['Time_spent_Alone', 'Social_event_attendance']]
y = data.Personality
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_test
```

```python
X_train
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score (X_test, y_test)
```

```python
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
```

```python
plt.plot(fpr, tpr, color='blue', label='ROC Curve')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
auc = roc_auc_score(y_test, y_pred_proba)
print("AUC Score:", auc)
```
